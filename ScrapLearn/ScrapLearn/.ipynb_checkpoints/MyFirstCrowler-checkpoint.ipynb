```python
# Set headers to mimic an actual browser
#Header passed to mimic an actual browser
headers = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'})
```

```python
prices = []
locations = []
sizes = []
descriptions = []
dates = []
urls = []
titles=[]
```

```python
import time
start = time.time()


#Set the variables which will be scrapped


n_pages = 0

for page in range(0,20):
    n_pages += 1
    
    sapo_url = 'https://casa.sapo.pt/Venda/Apartamentos/?sa=11&lp=10000&or=10&pn='+str(page)
    response = get(sapo_url, headers = headers)
    
    html = BeautifulSoup(response.text, 'html.parser')
    
    #house containers for the ith House scraped
    houseContainers = html.find_all('div', class_='searchResultProperty')
    if houseContainers != []:
        
        for container in houseContainers:
        
            #Get the container price
            price = container.find_all('span')[2].text
            #house price can return 'Contacte Anunciante', so we'll treat this
            if price == 'Contacte Anunciante':
                price = container.find_all('span')[3].text
                if price.find('/') != -1:
                    price = price[0:price.find('/')-1]
            if price.find('/') != -1:
                price = price[0:price.find('/')-1]
                
            if price == '':
                price = 0
            
            #Make it readable
            priceX = [int(price[s]) for s in range(0,len(price)) if price[s].isdigit()]
            price = ''

            for x in priceX:
                price = price+str(x)

            prices.append(float(price))
            
            #Get container location
            
            location = container.find_all('p', class_='searchPropertyLocation')[0].text
            #location = location[7:location.find(',')]
            location = location.replace("\r\n",'')
            locations.append(location)
            
            #Get container size
            
            size = container.find_all('p')[7].text
            size = (''.join(itertools.takewhile(str.isdigit,size)))
            sizes.append(size)
            
            #Get the day the container anounce was posted
            
            #date = pd.to_datetime(container.find_all('div', class_='searchPropertyDate')[0].text[21:31])
            #dates.append(date)
            
            #Get descriptions
            
            description = container.find_all('p', class_='searchPropertyDescription')[0].text
            description = description.replace('\r\n','')    
            descriptions.append(description)
            
            #Get urls
            
            url = ''
            for url in container.find_all('a'):
                #url.append(url.get('href'))
                #print(url.get('href'))
                url = url.get('href')
            urlx = 'https://casa.sapo.pt'+url
            urls.append(urlx)
            
            # Title
            name = container.find_all('span')[0].text
            name = name.replace('\r\n','') 
            titles.append(name)
            
    else:
        break
        
        sleep(randint(1,2))
        
        

            
print('You scraped {} pages containing {} properties.'.format(n_pages, len(titles)))
end = time.time()
print(end - start)
```

```python
cols = ['Title', 'Zone', 'Price', 'Size (m²)', 'Description', 'URL']

lisboa = pd.DataFrame({'Title': titles,
                           'Price': prices,
                           'Size (m²)': sizes,
                           'Zone': locations,
                           'Description': descriptions,
                           'URL': urls,})[cols]
```
